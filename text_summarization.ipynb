##  Prompt Template Text Summarization

```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    groq_api_key=os.environ["GROQ_API_KEY"]
)

template = """
Write a summary of the following speech:
Speech: {speech}

Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=template
)

chain = prompt | llm

response = chain.invoke({
    "speech": "This is a long speech about AI and the future.",
    "language": "French"
})

print(response.content)
```

```python
# Use this import for the latest version of LangChain
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the splitter
# Added "=200" to fix the red error line in your image
text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=200)

# To actually split your document (using the 'pages' from your PDF loader):
final_documents = text_splitter.split_documents(pages)

# Check how many chunks were created
print(f"Created {len(final_documents)} chunks.")
```

```python
from langchain_core.prompts import PromptTemplate

chunks_prompt="""
Please summarize the below speech:
Speech: `{text}`
Summary:
"""

map_prompt_template=PromptTemplate(input_variables=['text'],
                                   template=chunks_prompt)
```

```python
from langchain_core.prompts import PromptTemplate

final_prompt="""
Provide the final ummary of the entire speech with these important points.
Add a Motivation Title,Start the precise summary with an introduction and provide the summary in numbe
points for the speech.
Speech:{text}
"""

final_prompt_template = PromptTemplate(input_variables=['text'], template=final_prompt)
```

```python
# 1. Summarize each chunk individually (The "Map" step)
map_chain = map_prompt_template | llm
individual_summaries = map_chain.batch([{"text": d.page_content} for d in final_documents])

# 2. Join those summaries into one long string
combined_text = "\n\n".join([res.content for res in individual_summaries])

# 3. Create the final polished summary (The "Reduce" step)
reduce_chain = final_prompt_template | llm
final_output = reduce_chain.invoke({"text": combined_text})

# 4. Print the result
print(final_output.content)
```

##  Refine chain summarization

```python
from langchain.chains.summarize import load_summarize_chain

# Initialize the Refine chain
chain = load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)

# Run the chain on your split documents
# Use .invoke() if .run() gives you a deprecation warning
output_summary = chain.invoke(final_documents)

# Display the final refined summary
print(output_summary['output_text'])
```
